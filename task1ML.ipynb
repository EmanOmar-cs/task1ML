```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# قراءة الملف
df = pd.read_csv('titanic.csv')

# عرض أول 5 صفوف عشان تتأكدي إن الجدول ظهر
df.head()
```

```python
# كشف القيم المفقودة 
print("Missing values in each column:")
print(df.isnull().sum())

# رسم الـ Outliers لسعر التذكرة
sns.boxplot(x=df['Fare'])
plt.title('Outliers in Ticket Fare')
plt.show()
```

```python
# 1. كشف الصفوف المكررة (Duplicates)
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# 2. كشف عدم اتساق أنواع البيانات (Inconsistent Data Types)
# بنعرض نوع البيانات في كل عمود عشان نتأكد إن الأرقام متسجلة أرقام فعلاً
print("\nData types for each column:")
print(df.dtypes)
```
