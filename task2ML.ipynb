```python
import pandas as pd

# 1. تحميل البيانات
df = pd.read_csv('titanic.csv')

# 2. تحليل نسبة القيم المفقودة (Task 3 & 4)
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Missing Percentage per column:\n", missing_percentage)

# 3. معالجة القيم المفقودة (Imputation) - هنملا السن بالـ Median
df['Age'] = df['Age'].fillna(df['Age'].median())

# 4. الـ Encoding لعمود 'Sex' وحطه في نفس مكانه (Task 1 & 2)
col_idx = df.columns.get_loc('Sex')
df_encoded = pd.get_dummies(df['Sex'], prefix='Sex')

# حذف العمود القديم وإضافة الجديد مكانه
df.drop('Sex', axis=1, inplace=True)
for i, col in enumerate(df_encoded.columns):
    df.insert(col_idx + i, col, df_encoded[col])

# عرض النتيجة النهائية
print("\nFinal Data Preview:")
df.head()
```
